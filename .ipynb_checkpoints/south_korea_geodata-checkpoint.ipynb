## Downloading South Korea geodata through Natural Earth

At [Natural Earth](https://www.naturalearthdata.com/), you can download vector and raster map datasets at [different scales](https://www.naturalearthdata.com/downloads/).

[![ne_download_page](https://raw.githubusercontent.com/katkim0307/COVID-19_Response_Comparison/master/Data/image/ne_download_page.PNG)](https://www.naturalearthdata.com/downloads/)

## Download the zipfile and store it in the Data folder

```python
# import urllib

ne_admin1_states_url='https://www.naturalearthdata.com/http//www.naturalearthdata.com/download/10m/cultural/ne_10m_admin_1_states_provinces.zip'
```

```python
urllib.request.urlretrieve(ne_admin1_states_url, "Data/input/ne_admin_1_states_provinces.zip")
```

## Unzip the zipfile and Extract

```python
# import zipfile

ne_admin1_states_zip = zipfile.ZipFile('Data/input/ne_admin_1_states_provinces.zip')

ne_admin1_states_zip
```

```python
ne_admin1_states_zip.extractall('Data/input/ne_admin_1_states_provinces')
```

## Look in the directory and see what files are available

```python
# import glob

glob.glob('Data/input/ne_admin_1_states_provinces/*')
```

## Read the shapefile in this directory

```python
ne_admin_1_states_gdf = gpd.read_file('Data/input/ne_admin_1_states_provinces/ne_10m_admin_1_states_provinces.shp')

ne_admin_1_states_gdf.head()
```

```python
ne_admin_1_states_gdf.shape
```

```python
ne_admin_1_states_gdf.info()
```

## Filter out the gdf and get 41 admin='South Korea'

```python
south_korea_gdf=ne_admin_1_states_gdf[ne_admin_1_states_gdf['admin']=='South Korea']

south_korea_gdf
```

## Get the columns you need and make the dataframe easier to read

```python
south_korea_gdf = south_korea_gdf[
    ['geonunit', 'gu_a3', 'gn_name', 'name_de', 
     'fips', 'latitude', 'longitude', 'geometry']]

south_korea_gdf
```

## Plot South Korea

```python
south_korea_gdf.crs
```

```python
south_korea_gdf.plot(figsize=(8, 8));
```

## Let's add a title, legend, and basemap

```python
import contextily as ctx
```

```python
fig, ax = plt.subplots(figsize=(10, 15))

south_korea_gdf.plot(
    ax=ax,
    column='name_de',
    #cmap='Dark2',
    alpha=0.7,
    legend=True,
    legend_kwds={
        'loc':2,
        'shadow':True,
    }
)

# basemap
ctx.add_basemap(ax, source='https://{s}.basemaps.cartocdn.com/light_all/{z}/{x}/{y}.png')

ax.axis('off')

ax.set_title('South Korea States and Provinces', fontdict={'size': 20});
```
